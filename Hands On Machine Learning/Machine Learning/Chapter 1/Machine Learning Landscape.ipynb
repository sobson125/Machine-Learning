```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### loading data

```python
gdp_per_capita = pd.read_csv("../data/gdp_per_capita.csv", thousands=",", delimiter="\t", na_values="n/a", encoding='latin1')
oecd_2015 = pd.read_csv("../data/oecd_bli_2015.csv")
```

### perform initial transformations on

```python
oecd_2015 = oecd_2015[oecd_2015["INEQUALITY"]=="TOT"]
oecd_2015
```

```python
oecd_2015 = oecd_2015.pivot(index="Country", columns="Indicator", values="Value")
oecd_2015
```

```python
gdp_per_capita.rename(columns={"2015": "GDP per capita"}, inplace=True)
gdp_per_capita
```

```python
gdp_per_capita.set_index("Country", inplace=True)
gdp_per_capita
```

### Merge 2 dataframes into one

```python
full_country_stats = pd.merge(left=oecd_2015, right=gdp_per_capita,
                                  left_index=True, right_index=True)
full_country_stats.sort_values(by="GDP per capita", inplace=True)
remove_indices = [0, 1, 6, 8, 33, 34, 35]
keep_indices = list(set(range(36)) - set(remove_indices))
full_country_stats = full_country_stats[["GDP per capita", 'Life satisfaction']].iloc[keep_indices]
full_country_stats
```

```python
X = full_country_stats["GDP per capita"]
y = full_country_stats["Life satisfaction"]
```

```python
X
X = X.reset_index()
X.drop(["Country"],axis=1, inplace=True)
X
```

```python
y
```

### How the data looks on the plot

```python
full_country_stats.plot(kind='scatter', x="GDP per capita", y='Life satisfaction')
plt.show()
```

### Train model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

### Make predictions based on trained model

```python
X_new = [[22587]]  # GDP Cyprus
print(model.predict(X_new))
```
